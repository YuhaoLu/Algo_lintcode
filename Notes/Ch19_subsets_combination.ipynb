```python
# [65] Median of two Sorted Arrays
# There are two sorted arrays A and B of size m and n respectively. Find the median of the two sorted arrays.
# The overall run time complexity should be O(log(m + n))O(log(m+n)).

# Input:
# A = [1,2,3,4,5,6]
# B = [2,3,4,5]
# Output:
# 3.5

class Solution:
    def findMedianSortedArrays(self, A, B):
        # write your code here
        len_a, len_b = len(A), len(B)
        if (len_a + len_b) % 2 == 1:
            return self.finf_kth(A, B, (len_a + len_b) // 2 + 1)
        else:
            left = self.finf_kth(A, B, (len_a + len_b) // 2 )
            right = self.finf_kth(A, B, (len_a + len_b) // 2 + 1)
            return (left + right) / 2
            
    def finf_kth(self, A, B, k):
        if len(A) == 0:
            left, right = B[0], B[-1]
        elif len(B) == 0:
            left, right = A[0], A[-1]
        else:
            left, right = min(A[0], B[0]), max(A[-1], B[-1])
        while left + 1 < right:
            mid = (left + right) // 2
            count1 = self.helper(A, mid)
            count2 = self.helper(B, mid)
            if count1 + count2 < k:
                left = mid
            else:
                right = mid
        count1 = self.helper(A, left)
        count2 = self.helper(B, left)
        if count1 + count2 >= k:
            return left
        else:
            return right
    
    def helper(self, array, flag):
        if len(array) == 0:
            return 0
        left, right = 0 ,len(array) - 1
        while left + 1 < right:
            mid = (left + right) // 2
            if array[mid] <= flag:
                left = mid
            else:
                right = mid
        if array[right] <= flag:
            return right + 1
        if array[left] <= flag:
            return left + 1
        return 0
```

```python
# [17]  Subsets
# Given a set with distinct integers, return all possible subsets.
# Example 1:
# Input:
# nums = [1,2,3] 
# Output:
# [ 
#   [3], 
#   [1], 
#   [2], 
#   [1,2,3], 
#   [1,3], 
#   [2,3], 
#   [1,2], 
#   [] 
# ] 
# 
# Sol:
# 使用DFS组合的思路，对于每一位，考虑是否使用，是append再DFS，否pop再DFS
# 使用DFS排列的思路，对于每一位，从这一位往后进行DFS，DFS结束后要pop（回溯）
class Solution:
    def subsets(self, nums):
        results = []
        subset = []
        if nums is None:
            return results
        nums = sorted(nums)
        self.dfs(nums, 0, subset, results)
        return results

    # 1. 递归的定义
    def dfs(self, nums, index, subset, results):

        # 2. 递归的出口
        if index == nums.length:
            results.add(subset.copy())
            return

        # 3. 递归的拆解
        # 选 nums[index]
        subset.append(nums[index])
        self.dfs(nums, index+1, subset, results)

        # 不选 nums[index]
        subset.pop()
        self.dfs(nums, index+1, subset, results)
    
    # 
    #        ..[ ]..
    #      /    |    \
    #    [1]    [2]   [3]
    #   /   \     \
    # [1,2] [1,3] [2,3]
    # 
    def dfs2(self, nums, index, subset, results):
        results.append(subset.copy())
        
        for i in range(index, len(nums)):
            # [1] => [1, 2]
            # 去寻找以 [1, 2] 开头的所有子集
            subset.append(nums[i])
            self.dfs(nums, i + 1, subset, results)
            # [1, 2] => [1] Backtracking
        subset.pop()
sol = Solution()
nums = [1, 2, 3]
sol.subsets(nums)
```

```python
# [18] Subsets II
# Given a collection of integers that might contain duplicate numbers, return all possible subsets

# Example:
# Input:
# nums = [1,2,2]
# Output:
# [
#   [2],
#   [1],
#   [1,2,2],
#   [2,2],
#   [1,2],
#   []
# ]
class Solution:

    class Solution(object):
        def subsetsWithDup(self, nums):
            subset = []
            results = []
            nums.sort()
            self.dfs(nums, subset, results)
            return self.ans
        
        def dfs(self, nums, subset, results):
            results.append(subset)
            for i, n in enumerate(nums):
                if i > 0 and n == nums[i-1]:
                    continue
                self.dfs(nums[i+1:], subset + [n])
```
